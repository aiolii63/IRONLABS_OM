#### Now you will have to plot both $y=x^2$ and $y=e^x$ in the same plot. Create two subplots and use a normal scale in the first one and a logarithmic scale in the second one.
**Hint**: Use `set_xscale` and `set_yscale`.

```python
# your code here

x = np.arange(0.01, 10.0, 0.01)
data1 = x**2
data2 = np.exp(x)

fig, ax1 = plt.subplots()

color = 'tab:red'
ax1.set_xlabel('time')
ax1.set_ylabel('square', color=color)
ax1.plot(x, data1, color=color)
ax1.tick_params(axis='y', labelcolor=color)

ax2 = ax1.twinx()  # instantiate a second axes that shares the same x-axis

color = 'tab:blue'
ax2.set_ylabel('exp', color=color)  # we already handled the x-label with ax1
ax2.plot(x, data2, color=color)
ax2.tick_params(axis='y', labelcolor=color)

fig.tight_layout()  # otherwise the right y-label is slightly clipped
plt.show()
```

#### As a bonus challenge, try to add a legend to the plot.

```python
# your code here

x = np.arange(0.01, 10.0, 0.01)
data1 = x**2
data2 = np.exp(x)

fig, ax1 = plt.subplots()

color = 'tab:red'
ax1.set_xlabel('time')
ax1.set_ylabel('square', color=color)
ax1.plot(x, data1, color=color)
ax1.tick_params(axis='y', labelcolor=color)
plt.legend(["Square"], prop={"size":8})


ax2 = ax1.twinx()  # instantiate a second axes that shares the same x-axis

color = 'tab:blue'
ax2.set_ylabel('exp', color=color)  # we already handled the x-label with ax1
ax2.plot(x, data2, color=color)
ax2.tick_params(axis='y', labelcolor=color)
plt.legend(["Exp"], prop={"size":8})

fig.tight_layout()  # otherwise the right y-label is slightly clipped
plt.show()
```

```python
# MEDIAN STEPS BY MONTH_ENCODED
# your code here

df["Median Steps per month"] = df["Steps"].groupby(df['Months_encoded']).median()


# MEDIAN STEPS BY WORK_OR_WEEKEND
# your code here

df["Median Steps workday weekend"] = df["Steps"].groupby(df['Work_or_Weekend']).median().sort_values()
df

```

```python
# Find the median steps for each month
monthly_steps = df['Steps'].groupby(df['Months_encoded']).median()

# Plot the weekday steps in the current axes
monthly_steps.plot(kind = 'line',linewidth=4)

# Add labels, title, etc
plt.ylabel('Median number of steps')
plt.xlabel('Months')
plt.title('Median Steps walked by month')
plt.xticks(range(13), ['','Jan','Feb','Mar','Apr','May','Jun','Jul','Aug','Sep','Oct','Nov','Dec'])

```

```python
# Find the median steps for each weekday
daily_steps = df['Steps'].groupby(df['Days_encoded']).median().sort_values()

# Plot the weekday steps in the current axes
daily_steps.plot(kind = 'bar',linewidth=3)

# Add labels, title, etc
plt.ylabel('Median number of steps')
plt.xlabel('Days')
plt.title('Median Steps walked by day')
plt.xticks(range(7), ['Mon','Tue','Wed','Thu','Fri','Sat','Sun'])
```

```python
fig, [ax1, ax2] = plt.subplots(1,2, figsize=(15, 5))

plt.sca(ax1)

monthly_steps.plot(kind = 'line',linewidth=4)
plt.title('Median Steps walked by month')
plt.xlabel('Months')
plt.ylabel('Median number of steps')
plt.xticks(range(13), ['','Jan','Feb','Mar','Apr','May','Jun','Jul','Aug','Sep','Oct','Nov','Dec'])

plt.sca(ax2)
daily_steps.plot(kind = 'bar',linewidth=4)
plt.title('Median number of steps')
plt.xlabel('Days')
plt.ylabel('Median number of steps')
plt.xticks(range(7), ['Mon','Tue','Wed','Thu','Fri','Sat','Sun'])

plt.show()

```

#### Plot the correlation matrix using seaborn.

```python
# your code here

from string import ascii_letters
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

sns.set_theme(style="white")

# Generate a large random dataset
df = data


# Compute the correlation matrix
corr = df.corr()

# Generate a mask for the upper triangle
mask = np.triu(np.ones_like(corr, dtype=bool))

# Set up the matplotlib figure
f, ax = plt.subplots(figsize=(11, 9))

# Generate a custom diverging colormap
cmap = sns.diverging_palette(230, 20, as_cmap=True)

# Draw the heatmap with the mask and correct aspect ratio
sns.heatmap(corr, mask=mask, cmap=cmap, vmax=.3, center=0,
            square=True, linewidths=.5, cbar_kws={"shrink": .5})
```
