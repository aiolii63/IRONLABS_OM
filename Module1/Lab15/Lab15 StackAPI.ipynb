```python
#Question 1: Find the questions and answers of last month.

from stackapi import StackAPI
SITE = StackAPI('stackoverflow')
questions = SITE.fetch('questions', fromdate=1646151388, todate=1646151702)
questions
```

```python
from stackapi import StackAPI
SITE = StackAPI('stackoverflow')
answers = SITE.fetch('answers', fromdate=1646151388, todate=1646151702)
answers
```

```python
#Question 2: Find the most voted question today with at least a score of 5 and tagged with 'python'.
```

```python
SITE = StackAPI('stackoverflow')
questions = SITE.fetch('questions', fromdate=1646611201, todate=1646697599,
    min=5, tagged='python', sort='votes')
questions
```
